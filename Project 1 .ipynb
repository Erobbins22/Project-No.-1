```python
#### Show you have imported the pandas package in the space below
import pandas as pd

air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')

air_data = pd.DataFrame(air_data)


```

### Step 2: This function prints the location with the highest mean and median concentrations of VOC, PM2.5, and PM10.0. This will also include the data for each of these data sets. 
### (Task 1 & 2)

This specific table below is for VOC values, with the respective locations.

```python

air_data_grouped = air_data.groupby(['date', 'sensor.name'])
air_data_voc = air_data_grouped.agg(mean = ('voc', 'mean'),
                                    median = ('voc', 'median'))

air_data_voc

top5_mean_voc = (
    air_data_voc.sort_values(by = 'mean', ascending = False)
)
top5_mean_voc.head(5)
```

This specific table below is for PM2.5 values, with the respective locations and dates the values were taken.

```python
air_data_grouped = air_data.groupby(['date','sensor.name'])
air_data_pm25 = air_data_grouped.agg(mean = ('pm2.5_atm', 'mean'),
                                     median = ('pm2.5_atm', 'median'))

air_data_pm25

top5_mean_pm25 = (
    air_data_pm25.sort_values(by = 'mean', ascending = False)
)
top5_mean_pm25.head(5)
```

This specific table below is for PM1.0 values, with the respective locations and dates for the data. 

```python
### Print your dataframe in this space

air_data_grouped = air_data.groupby(['date','sensor.name'])
air_data_pm10 = air_data_grouped.agg(mean = ('pm10.0_atm', 'mean'),
                                     median = ('pm10.0_atm', 'median'))

air_data_pm10

top5_mean_pm10 = (
    air_data_pm10.sort_values(by = 'mean', ascending = False)
)
top5_mean_pm10.head(5)

```

### Temperature and Humidity Effecting the Air Quality (Task 3)



```python
def assign_temperature(temperature):
    if temperature <32:
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 50 <= temperature <= 70:
        return 'Warm'
    else:
        return 'Hot'
air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature) 
#the input to my assign_temperature function would be the data column temperature 
air_data.head(5)
```

```python
def assign_humidity(X):
    if X <50:
        return 'Low'
    elif 50 <= X <= 80:
        return 'High'
    else:
        return 'Very High'

air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity) 
#the input to my assign_temperature function would be the data column temperature 
air_data.tail(30)
```

Based on the data above, the temperature does affect air quality, with higher temperatures corresponding with increased VOCs values, and in certain locations, it raises PM2.5 and PM10 values. The humidity doesn't have as strong an influence on the air quality; the humidity levels remained low and didn't show much variation. Overall, the temperature is a larger factor affecting air quality than humidity for the time period and locations presented. 

### Unhealthy Values (Task 4)

The tables below present values for the PM2.5 and PM10.0 that risk being unhealthy for sensitive groups. 

```python
unhealthy_pm25 = air_data[
    (air_data['pm2.5_atm']>=35.5) &
    (air_data['pm2.5_atm']<=55.4)]
unhealthy_events = unhealthy_pm25[
    ['date', 'sensor.name', 'pm2.5_atm']
].sort_values('date')
unhealthy_events

```

During the winter, the temperature inversions increase because a layer of warm air can trap cold air and the pollutants within it close to the ground. During the holidays, as well, there is an increase in wood-burning for heating homes. This and the combinations of stagnant air lead to spikes in PM2.5.
